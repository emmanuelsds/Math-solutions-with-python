```python
# Import Libraries
import matplotlib.pyplot as plt
```

```python
#Get User Input for Expenditures
def get_expenditures(num_categories):
    categories = []
    expenditures = []
    for i in range(num_categories):
        category = input(f"Enter category {i+1}: ")
        expenditure = float(input("Expenditure: "))
        categories.append(category)
        expenditures.append(expenditure)
    return categories, expenditures
```

```python
#Categories and expenditures
def plot_bar_chart(categories, expenditures):
    plt.figure(figsize=(10, 5))
    plt.bar(categories, expenditures, color='skyblue')
    plt.xlabel('Categories')
    plt.ylabel('Weekly Expenditure ($)')
    plt.title('Weekly Expenditures by Category')
    plt.show()
```

```python
#Main Function
def main():
    try:
        num_categories = int(input("Enter the number of categories: "))
        if num_categories <= 0:
            raise ValueError("Number of categories must be a positive integer.")
        
        categories, expenditures = get_expenditures(num_categories)
        plot_bar_chart(categories, expenditures)
    
    except ValueError as e:
        print(f"Error: {e}")
```

```python
if __name__ == "__main__":
    main()
```

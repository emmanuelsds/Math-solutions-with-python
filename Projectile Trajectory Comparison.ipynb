```python
#Import the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Constants
g = 9.81  # Acceleration due to gravity (m/s^2)
```

```python
#Calculate project motion
def projectile_motion(v0, theta):
    theta_rad = np.deg2rad(theta)
    t_flight = 2 * v0 * np.sin(theta_rad) / g
    max_horizontal = v0**2 * np.sin(2*theta_rad) / g
    max_vertical = (v0**2) * (np.sin(theta_rad)**2) / (2*g)
    return t_flight, max_horizontal, max_vertical
```

```python
#Plot trajectory
def plot_trajectory(v0, theta, t_flight):
    theta_rad = np.deg2rad(theta)
    t = np.linspace(0, t_flight, num=1000)
    x = v0 * t * np.cos(theta_rad)
    y = v0 * t * np.sin(theta_rad) - 0.5 * g * t**2
    plt.plot(x, y, label=f'v0={v0} m/s, θ={theta}°')
```

```python
#Get user input for trajectories
def get_user_input(num_trajectories):
    trajectories = []
    for i in range(1, num_trajectories + 1):
        v0 = float(input(f"Enter the initial velocity for trajectory {i} (m/s): "))
        theta = float(input(f"Enter the angle of projection for trajectory {i} (degrees): "))
        trajectories.append((v0, theta))
    return trajectories
```

```python
#Main funcion
def main():
    try:
        num_trajectories = int(input("How many trajectories? "))
        if num_trajectories <= 0:
            raise ValueError("Number of trajectories must be a positive integer.")
        
        trajectories = get_user_input(num_trajectories)
        
        for i, (v0, theta) in enumerate(trajectories, start=1):
            t_flight, max_horizontal, max_vertical = projectile_motion(v0, theta)
            print(f"\nTrajectory {i}:")
            print(f"Time of Flight: {t_flight:.2f} s")
            print(f"Maximum Horizontal Distance: {max_horizontal:.2f} m")
            print(f"Maximum Vertical Distance: {max_vertical:.2f} m")
            
            plot_trajectory(v0, theta, t_flight)
        
        plt.title('Projectile Motion of Different Trajectories')
        plt.xlabel('Horizontal Distance (m)')
        plt.ylabel('Vertical Distance (m)')
        plt.grid(True)
        plt.legend()
        plt.show()
    
    except ValueError as e:
        print(f"Error: {e}")
```

```python
#Run the program
if __name__ == "__main__":
    main()
```
